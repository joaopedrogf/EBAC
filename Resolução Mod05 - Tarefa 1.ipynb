````python
import pandas as pd
import requests
import numpy as np

# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
# Não há duplicados
````

````python
# 2) seu código aqui
sinasc.isna().sum()
````

````python
# 3) seu código aqui
sinasc[['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']].isna().sum()
````

````python
# 4) seu código aqui
sinasc['APGAR5'].dropna(inplace=True)
sinasc['APGAR5'].isna().sum()
````

````python
# 5) seu código aqui
sinasc.loc[sinasc.ESTCIVMAE==9, 'ESTCIVMAE'] = np.nan
sinasc.loc[sinasc.CONSULTAS==9, 'CONSULTAS'] = np.nan
sinasc.loc[(sinasc['ESTCIVMAE'] == 9) | (sinasc['CONSULTAS'] == 9)]
#Transformando o 9 como vazio.
````

````python
# 6) seu código aqui
sinasc['QTDFILVIVO'].fillna(0, inplace=True)
sinasc['QTDFILVIVO'].isna().sum()
````

````python
# 7) seu código aqui
sinasc['GRAVIDEZ'].fillna(0, inplace=True)
# Decidi preencher os valores nulos da coluna "GRAVIDEZ" para 0. Faz sentido, porque se a pessoa não preencheu, provavelmente ela nunca engravidou.
````

````python
# 8) seu código aqui
sinasc['APGAR'] = sinasc['APGAR5'].map({10:'normal', 9:'normal',
                           8:'normal',7:'asfixia leve',
                           6:'asfixia leve',5:'asfixia moderada',
                           4:'asfixia moderada',3:'asfixia severa',
                           2:'asfixia severa',1:'asfixia severa',0:'asfixia severa'})
sinasc['APGAR'].value_counts()
````
