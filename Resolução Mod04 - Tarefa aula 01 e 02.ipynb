#### 5) Nomeie os índices das linhas com inteiros de 1 a 20, e as colunas com os nomes "x1", "x2", e "x3" respectivamente.

```python
#rename index
df_novo = df.rename(index={0:'x1', 1:'x2', 2:'x3', 3:'x4',4:'x5' ,5:'x6' ,6:'x7' ,7:'x8' ,8:'x9' ,9:'x10' ,10:'x11' ,11:'x12' ,12:'x13' ,13:'x14' ,14:'x15' ,15:'x16' ,16:'x17' ,17:'x18' ,18:'x19' ,19:'x20'})
df_novo
```

#### 6) No DataFrame do exercício 5, crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media" (não recomendo colocar acentos em nomes de variáveis).

```python
# para encontrarmos a média de uma linha, basta utilizarmos a função built-in "mean()" do python passando como parâmetro o eixo de busca
df_novo['media'] = df_novo.mean(axis=1)
df_novo
```

#### 7) No DataFrame do exercício 6, crie uma nova coluna chamada "log_med", contendo o logaritmo natural da média calculada no exercício 6 <br>

```python
# utilizaremos o np.log (função da biblioteca numpy para encontrar o logaritmo natural da média)
df_novo['log_med'] = np.log(df_novo['media'])
df_novo
```
